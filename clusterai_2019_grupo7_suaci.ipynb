```python
base3 = base3.drop('domicilio_esquina_proxima', 1)
base2 = base2.drop('domicilio_esquina_proxima', 1)
```

```python
base2[base2['estado_del_contacto'] == 'Abierto']
```

```python
# validamos que los registros con 'estado_del_contacto' == 'Abierto' presentan valor nulo en el campo 'fecha_cierre_contacto'
base2_estado_contacto = base2.groupby('estado_del_contacto')['categoria'].count()
base2_estado_contacto
```

```python
# creamos el campo 'plazo_resolución', definido como el plazo entre las fechas de apertura y cierre de contacto
base2['fecha_cierre_contacto'] = pd.to_datetime(base2['fecha_cierre_contacto'])
base2['fecha_ingreso'] = pd.to_datetime(base2['fecha_ingreso'])
base2['plazo_resolucion'] = base2['fecha_cierre_contacto'] - base2['fecha_ingreso']
base2['plazo_resolucion'] = base2['plazo_resolucion'].dt.days
```

```python
# redefinición valor 'domicilio_cgpc' para su ordenamiento
base3['domicilio_cgpc'] = base3['domicilio_cgpc'].replace({'COMUNA 1':'COMUNA 01', 'COMUNA 2':'COMUNA 02', 'COMUNA 3':'COMUNA 03', 'COMUNA 4':'COMUNA 04', 'COMUNA 5':'COMUNA 05', 'COMUNA 6':'COMUNA 06', 'COMUNA 7':'COMUNA 07', 'COMUNA 8':'COMUNA 08', 'COMUNA 9':'COMUNA 09'})
```

```python
base3_comunas = sorted(set(base3['domicilio_cgpc'].unique()))
base3_comunas
```

A continuación presentamos algunas visualizaciones de la información agrupada.

```python
# countplot: cantidad de contactos por comuna
base3['domicilio_cgpc'] = pd.Categorical(base3['domicilio_cgpc'], categories = base3_comunas , ordered = True)
plt.figure(figsize=(10,6))
ax = sns.countplot(x='domicilio_cgpc', data=base3)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.title("Cantidad de contactos al SUACI por CGPC", fontsize=16)
plt.show()
```

```python
# evolución temporal de los contactos
base3['fecha_ingreso'] = pd.to_datetime(base3['fecha_ingreso'])
base3['anio'] = base3['fecha_ingreso'].dt.year
base3['mes'] = base3['fecha_ingreso'].dt.month
base3_contactos_periodo = pd.pivot_table(base3, values = 'canal', index = 'mes', columns = 'anio', aggfunc = 'count')
plt.figure(figsize=(10,6))
plt.plot(base3_contactos_periodo.index, base3_contactos_periodo[2017], label='2017');
plt.plot(base3_contactos_periodo.index, base3_contactos_periodo[2018], label='2018');
plt.plot(base3_contactos_periodo.index, base3_contactos_periodo[2019], label='2019');
plt.legend(loc='upper right')
plt.title('Evolución temporal de los contactos al SUACI', fontsize=16)
plt.ylabel('Cantidad de contactos por periodo')
plt.xlabel('Mes')
plt.show()
```

```python
# estandarización categorías
np.sort(base3['categoria'].unique())
```

```python
base3['categoria'] = base3['categoria'].replace({'FISCALIZACIÓN ACTIVIDADES COMERCIALES':'FISCALIZACION ACTIVIDADES COMERCIALES', 'LIMPIEZA Y RECOLECCIÓN':'LIMPIEZA Y RECOLECCION', 'ORDENAMIENTO DEL ESPACIO PÚBLICO':'ORDENAMIENTO DEL ESPACIO PUBLICO', 'RECICLADO Y PROTECCIÓN AMBIENTAL':'RECICLADO Y PROTECCION AMBIENTAL', 'TRÁMITES Y SERVICIOS':'TRAMITES Y SERVICIOS', 'TRÁNSITO':'TRANSITO'})
```

```python
# total de contactos por categoría
base3_contactos_categoria = base3.groupby('categoria')['canal'].count()
base3_contactos_categoria
```

```python
# participación porcentual de cada categoría en el total de contactos registrados
base3_contactos_categoria_pcts = base3_contactos_categoria.apply(lambda x: 100 * x / np.shape(base3)[0])
base3_contactos_categoria_pcts.sort_values(ascending = False)
```

```python
# relación entre categorías
base3_categoria_periodo = pd.pivot_table(base3, values = 'canal', index = 'periodo', columns = 'categoria', aggfunc = 'count')
plt.figure(figsize=(12,8))
sns.heatmap(base3_categoria_periodo)
plt.title("Cantidad de contactos por mes y categoría", fontsize=16)
plt.show()
```

Algunas observaciones:
- Medio ambiente y reciclado presenta registros sólo en 2017. ¿Se dejó de informar esta categoría, o bien sus contactos se reclasificaron dentro de otra?
- Cementerios está actualizada hasta junio-2019
- Tenemos 2 categorías preponderantes ('Limpieza y recolección' / 'Tránsito') que explican > 60% de los contactos. 5 categorías (las anteriores + 'Arbolado y espacios verdes' / 'Alumbrado' / 'Calles y veredas') responden por más del 80%.

```python
# análisis de correlación lineal entre las principales categorías
plt.figure(figsize=(10,6))
sns.heatmap(base3_categoria_periodo.loc[:, base3_contactos_categoria_pcts.head(5).index].corr())
plt.title("Correlacion entre las principales categorias por mes", fontsize=16)
plt.show()
```

```python
# relación entre comunas
base3_comunas_periodo = pd.pivot_table(base3, values = 'canal', index = 'periodo', columns = 'domicilio_cgpc', aggfunc = 'count')
plt.figure(figsize=(12,8))
sns.heatmap(base3_comunas_periodo)
plt.title("Cantidad de contactos por mes y comuna", fontsize=16)
plt.show()
```

```python
# análisis de correlación lineal entre las comunas
plt.figure(figsize=(10,6))
sns.heatmap(base3_comunas_periodo.corr())
plt.title("Correlacion entre las comunas por mes", fontsize=16)
plt.show()
```

```python
# una última iteración en busca de relaciones: comunas vs categorias (total contactos)
base3_comunas_categorias = pd.pivot_table(base3, values = 'canal', index = 'categoria', columns = 'domicilio_cgpc', aggfunc = 'count')
plt.figure(figsize=(12,8))
sns.heatmap(base3_comunas_categorias)
plt.title("Cantidad de contactos por categoría y comuna", fontsize=16)
plt.show()
```

```python
# proporción de contactos abiertos/cerrados
base3['estado_del_contacto'] = base3['estado_del_contacto'].replace({'ABIERTO':'Abierto', 'CERRADO':'Cerrado'})
base3_estado_contactos = base3.groupby(['anio', 'estado_del_contacto'])['canal'].count()
base3_estado_contactos_pcts = base3_estado_contactos.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
base3_estado_contactos_pcts
```

Sobre esta última tabla: al estar el ejercicio 2019 en curso, cabría esperar una mayor proporción de contactos abiertos. Sin embargo, las cifras muestran valores similares a los del año anterior.

```python
# grafico proporción de contactos abiertos/cerrados
plt.figure(figsize=(10,6))
ax = sns.countplot(x='anio', data=base3, hue='estado_del_contacto')
ax.set_xticklabels(ax.get_xticklabels(), ha="right")
plt.title("Estado de los contactos al SUACI por año", fontsize = 16)
plt.show()
```
